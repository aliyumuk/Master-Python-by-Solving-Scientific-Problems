```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
population_size = 230000
skip = 1000
sample = 1 / np.logspace(np.log10(0.001),np.log10(10),int(population_size/skip))


plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
plt.plot(sample, 'o')

plt.subplot(1,2,2)
np.random.shuffle(sample)
plt.plot(sample, 'o',)
plt.show()

true_mean = np.mean(sample)
print("True mean: ", true_mean)
```

```python
num_of_samples = 50
num_of_experiments = 500

averages = np.zeros(num_of_experiments)
for j in range(num_of_experiments):
    exp_mean = 0
    for i in range(num_of_samples):
        exp_mean += sample[np.random.randint(population_size/skip)]
    exp_mean /= num_of_samples
    averages[j] = exp_mean


plt.plot(averages, 'ko', label ='Sample Means')
plt.plot([0, 500], [np.mean(averages), np.mean(averages)], label='Calculated Average', linewidth=5)
plt.legend()
plt.xlabel("Sample Number")
plt.ylabel("Mean Value")
plt.show()

print("Calculated mean: ", np.mean(averages))
```

```python
cumulative_averages = np.zeros(num_of_experiments)
cumulative_averages[0] = averages[0]
for i in range(1,num_of_experiments):
    cumulative_averages[i] = np.mean(averages[:i])

plt.plot(cumulative_averages)
plt.plot([0, 500], [np.mean(averages), np.mean(averages)],
         label='Calculated Average', linewidth=3)
plt.legend()
plt.xlabel("Sample Number")
plt.ylabel("Mean Value")
plt.show()
```
