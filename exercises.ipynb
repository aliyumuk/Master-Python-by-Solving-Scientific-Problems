```python
n = 0
for item in "abcdefghijklmnopqrstuvwxyz":
    n += 1
    th = "th"
    if n == 1 or n == 21: th = "st"
    if n == 2 or n == 22: th = "nd"
    if n == 3 or n == 23: th = "rd"
    #print(item + " is the " + str(n) + th + " letter of the alphabet." )
    print("%s is the %d%s elemeth of the alphabet."%(item,n,th))
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

n = 100
hilbert = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        hilbert[i][j] = 1/(i+1 + j+1 - 1)

print(np.round(hilbert,2))
plt.subplot(2,1,1)
sns.heatmap(hilbert, vmin=0, vmax=0.02)
plt.subplot(2, 1, 2)
sns.heatmap(hilbert, vmin=0, vmax=0.5)
plt.show()
```
